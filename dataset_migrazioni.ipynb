```python
df_migr_pop_unito = pd.concat([df_migr_pop_lt5, df_migr_pop_5_9, df_migr_pop_10_14, df_migr_pop_15_19, df_migr_pop_20_24, df_migr_pop_25_29, df_migr_pop_30_34, df_migr_pop_35_39, df_migr_pop_40_44, df_migr_pop_45_49, df_migr_pop_50_54, df_migr_pop_55_59, df_migr_pop_60_64, df_migr_pop_65_69, df_migr_pop_70_74, df_migr_pop_75_79, df_migr_pop_80_84, df_migr_pop_gt85], ignore_index=True)
```

```python
df_migr_pop_unito
```

```python
# Conta i valori nulli per ciascuna riga
null_counts_per_row = df_migr_pop_unito.isnull().sum(axis=1)

# Filtra le righe che hanno da 1 a 2 valori nulli
few_null_pop = df_migr_pop_unito[(null_counts_per_row >= 1) & (null_counts_per_row <= 2)]

# Visualizza le righe selezionate
few_null_pop
```

```python
for index in few_null_pop.index:
    # Seleziona le colonne numeriche escluse le colonne 'country', 'sex', e 'age_category'
    valori_numerici = df_migr_pop_unito.loc[index, df_migr_pop_unito.columns.difference(['country', 'sex', 'age category'])]
    
    # Converti in numerico, ignorando valori non numerici
    valori_numerici = pd.to_numeric(valori_numerici, errors='coerce')
    
    # Stampa i valori usati nel calcolo della media per fare debug
    print(f"Valori per il calcolo della media all'indice {index}:\n", valori_numerici)
    
    # Calcola la media sui valori numerici effettivi, escludendo i NaN
    mean_value_pop = valori_numerici.mean(skipna=True)
    print(f"Media calcolata all'indice {index}: {mean_value_pop}")  # Per debug

    # Riempie i NaN nella riga con la media calcolata
    df_migr_pop_unito.loc[index] = df_migr_pop_unito.loc[index].fillna(round(mean_value_pop, 1))
```
